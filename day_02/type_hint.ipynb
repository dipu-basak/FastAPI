```python
def func(name: str | None = None):
    if name is not None:
        print(f"Hello, {name}")
    else:
      print("hello world")

func('Basak')
```

```python
# now without any argument
def func(name: str | None = None):
    if name is not None:
        print(f"Hello, {name}")
    else:
      print("hello world")

func()
```

```python
# None = None assigns a default value to the parameter.
# None = 2 or None = "Bruce Wayne" works too
# in case of absence of argument, default value is used
def func(name: str | None = "Bruce Wyane"):
    if name is not None:
        print(f"Hello, {name}")
    else:
      print("hello world")

func()
```

```python
# usnig typing
import typing

def func(name: typing.Optional[str] = None):
  if name is not None:
        print(f"Hello, {name}")
  else:
      print("hello world")

func()
```

```python
# list type
def func(items: list[str]):
  for item in items:
    print(item.capitalize())

func(['hi', 'hello', 'bye'])
```

```python
# using typing
from typing import List

def func(items: List[str]):
  for item in items:
    print(item.capitalize())

func(['hi', 'hello', 'bye'])
```
